```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pyreadr
import os
import numpy as np
```

```python
def load_rda(file_paths):

  dataframes = {}

  for file_path in file_paths:
      result = pyreadr.read_r(file_path)

      for key in result.keys():
          dataframes[key] = result[key]

  return dataframes
```

```python
def missing_values(df):
  return(df.isnull().sum())
```

```python
import pandas as pd

def outlier_values(df):
    outliers = pd.Series(dtype="float64")

    for col in df.select_dtypes(include=["number"]):  # S√©lectionner uniquement les colonnes num√©riques
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # D√©tection des outliers
        mask = (df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))
        outliers[col] = mask.mean() * 100  # Calcul du pourcentage

    return outliers
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_variable_distributions(dataframe, output_dir="plots"):
    """
    G√©n√®re et affiche les distributions des variables num√©riques et cat√©goriques 
    avec des couleurs distinctes pour histogrammes et courbes KDE.

    Param√®tres :
        dataframe (pd.DataFrame): Le DataFrame contenant les donn√©es.
        output_dir (str): R√©pertoire o√π sauvegarder les graphiques.
    """
    # Cr√©er le r√©pertoire si n√©cessaire
    os.makedirs(output_dir, exist_ok=True)

    # G√©n√©rer une palette de couleurs vari√©es
    num_colors = len(dataframe.select_dtypes(include='number').columns)
    colors = sns.color_palette("husl", n_colors=num_colors)

    # üìä Variables num√©riques
    for i, column in enumerate(dataframe.select_dtypes(include='number').columns):
        plt.figure(figsize=(10, 6))
        sns.histplot(dataframe[column], kde=True, 
                     color=colors[i], edgecolor="black", alpha=0.6,  # Histogramme
                     line_kws={"linewidth": 2})  # Courbe KDE
        plt.title(f"Distribution de {column}")
        plt.xlabel(column)
        plt.ylabel("Fr√©quence")
        plt.show()  # Affichage direct dans Jupyter

    # üìä Variables cat√©goriques
    for column in dataframe.select_dtypes(include='category').columns:
        plt.figure(figsize=(10, 6))
        sns.countplot(y=dataframe[column], order=dataframe[column].value_counts().index, palette="Set2")
        plt.title(f"Distribution de {column}")
        plt.xlabel("Fr√©quence")
        plt.ylabel(column)
        plt.show()  # Affichage direct dans Jupyter
```

```python
def plot_boxplots(df):
    numeric_columns = df.select_dtypes(include='number').columns
    plt.figure(figsize=(15, 10))

    # Cr√©er un boxplot pour chaque variable num√©rique
    for i, col in enumerate(numeric_columns, 1):
        plt.subplot(4, 4, i)
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot de {col}')

    plt.tight_layout()
    plt.show()
```

```python
def clean_data(data_dict):
    cleaned_dict = {}
    for key, df in data_dict.items():
        df_cleaned = df.drop_duplicates()
        if 'RecordBeg' in df_cleaned.columns:
            df_cleaned.loc[:, 'RecordBeg'] = pd.to_datetime(df_cleaned['RecordBeg'], errors='coerce')
        if 'RecordEnd' in df_cleaned.columns:
            df_cleaned.loc[:, 'RecordEnd'] = pd.to_datetime(df_cleaned['RecordEnd'], errors='coerce')

        cleaned_dict[key] = df_cleaned

    return cleaned_dict
```
